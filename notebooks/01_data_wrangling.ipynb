```python
# look at missing data
missing = pd.concat([sd_data_wave1.isnull().sum(), 100 * sd_data_wave1.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
print(missing)
```

```python
# filter out columns where missing values are 100%
missing_cols = missing[missing['count'] == 200].index
print(missing_cols)
```

```python
# drop columns with 100% missing data
sd_data_wave1.drop(missing_cols, axis=1, inplace=True)
sd_data_wave1.info()
```

```python
# look at missing data again
missing = pd.concat([sd_data_wave1.isnull().sum(), 100 * sd_data_wave1.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
missing
```

```python
# filter columns where missing values greater than 0
# potential problems for later
some_missing_cols = missing[missing['count'] > 0].index
print(some_missing_cols)
```
